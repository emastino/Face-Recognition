```python
# Successfully uninstalled numpy-1.19.4
# Successfully installed numpy-1.19.3 ( needed this version of numpy for it to work)
# import numpy
import numpy as np

# import open source coputer vision library
import cv2

import pickle

%matplotlib inline 

#import matplotlib.pyplot
import matplotlib.pyplot as plt

# import matplotlib.image
import matplotlib.image as mpimg
```

```python
# cascade only for front of face. Profile of face will hardly get recognized!!
face_cascade = cv2.CascadeClassifier('cascades/data/haarcascade_frontalface_alt2.xml')

# eye cascade for eyes
eye_cascade = cv2.CascadeClassifier('cascades/data/haarcascade_eye.xml')

# recognizer
recognizer = cv2.face.LBPHFaceRecognizer_create()
recognizer.read("trainner.yml")

labels = {"person_name": 1}
with open("labels.pickle",'rb') as f:
    og_labels = pickle.load(f)
    labels = {v:k for k, v in og_labels.items()}

    
print(labels)
# video capture with primary webcam
cap = cv2.VideoCapture(0)
# cap.release()
# cv2.destroyAllWindows()        
```

```python


# video loop
while True:
    ret, frame = cap.read()
    gray = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, scaleFactor =1.5, minNeighbors =5)
    
    for (x,y,w,h) in faces:
#         print(x,y,w,h)
        offset = -20
        # region of interest in gray from y to y+h and x to x+w
        roi_gray = gray[y-offset:y+h+offset,x-offset:x+w+offset]
        
        # region of interest color from y to y+h and x to x+w
        roi_color = frame[y-offset:y+h+offset,x-offset:x+w+offset]
        
        #recognizer
        id_ ,conf = recognizer.predict(roi_gray)
        
        print(labels[id_], conf)
        # save image
        img_item = "11.jpg" # image name
        cv2.imwrite(img_item,frame) # save image
        
        # make box
        offset = 20
        boxColor = (25,200, 50) # BGR why!? from 0-255
        boxLineWidth = 3
        cv2.rectangle(frame, (x-offset,y-offset), (x+w+offset,y+h+offset), boxColor, boxLineWidth)
        
        
        # only put name if confident in prediction
        if conf >40:
            # add text to box
            font = cv2.FONT_ITALIC
            fontScale = 1
            cv2.putText(frame, labels[id_], (x,y+h), font,  
                       fontScale, boxColor, boxLineWidth, cv2.LINE_AA)
            
            
        eyes = eye_cascade.detectMultiScale(roi_gray)
        
        for (ex,ey,ew,eh) in eyes:
            cv2.rectangle(roi_color, (ex,ey), (ex+ew,ey+eh), (255,0,0), 1)
        
    cv2.imshow('frame', frame)
    
    if cv2.waitKey(20) & 0xFF == ord('q'):
        break
        
cap.release()
cv2.destroyAllWindows()
```
